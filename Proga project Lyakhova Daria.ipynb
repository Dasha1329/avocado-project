```python
data = data.sort_values(['Date'])

sns.scatterplot(x='Date', y='Total Volume', data = data, color='darkorchid')
```

```python
mean_Average_Price = round(data['AveragePrice'].mean(), 1)
mean_Total_Volume = round(data['Total Volume'].mean(), 1)
mean_Total_Bags = round(data['Total Bags'].mean(), 1)

median_Average_Price = data['AveragePrice'].median()  
median_Total_Volume = data['Total Volume'].median()
median_Total_Bags = data['Total Bags'].median()

std_Average_Price = round(data['AveragePrice'].std(), 1) 
std_Total_Volume = round(data['Total Volume'].std(), 1)
std_Total_Bags = round(data['Total Bags'].std(), 1)
```

```python
data_description = pd.DataFrame({'Type' : ['Average_Price', 'Total_Volume', 'Total_Bags']})
data_description ['Mean'] = [mean_Average_Price, mean_Total_Volume, mean_Total_Bags]
data_description ['Median'] = [median_Average_Price, median_Total_Volume, median_Total_Bags]
data_description ['Standart deviation'] = [std_Average_Price, std_Total_Volume, std_Total_Bags]
data_description
```

```python
sns.scatterplot(x = 'Date', y = 'Total Bags', data = data, color = 'palevioletred')
```

```python
import seaborn as sns

sns.heatmap(
    data.assign(
        Date = lambda data: pd.to_datetime(data.Date),
        year = lambda data: data.Date.dt.year,
        month = lambda data: data.Date.dt.month
    ).pivot_table(
        index='year',
        columns='month',
        values='AveragePrice',
        aggfunc='mean',
        fill_value=0
    )
)
```
